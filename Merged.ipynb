```python
df['A_X'] = df['A_X'].apply(lambda x:float(x))
df['A_Y'] = df['A_Y'].apply(lambda x:float(x))
df['A_Z'] = df['A_Z'].apply(lambda x:float(x))
df['Time'] = df['Time'].apply(lambda x:float(x))
```

```python
import numpy as np
import pandas as pd
from scipy.signal import butter, filtfilt




# Define the parameters for the Butterworth bandpass filter
fs = 25 # sample rate
lowcut = 0.5 # low cutoff frequency
highcut = 125 # high cutoff frequency
order = 5 # filter order

# Normalize the cutoff frequencies to the Nyquist frequency
nyquist = 0.5 * fs
low = lowcut / nyquist
#high = highcut / nyquist

# Calculate the Butterworth bandpass filter coefficients
b, a = butter(order, low, btype='low')

# Apply the Butterworth bandpass filter to the PPG data
A_X = df['A_X'].values
A_Y = df['A_Y'].values
A_Z = df['A_Z'].values
filtered_X = filtfilt(b, a, A_X)
filtered_Y = filtfilt(b, a, A_Y)
filtered_Z = filtfilt(b, a, A_Z)

# Create a new DataFrame for the filtered data
dfm1 = pd.DataFrame({'Time': df['Time'], 'A_X': filtered_X,'A_Y':filtered_Y,'A_Z':filtered_Z, 'Activity': df['Activity']})
dfm1
```
